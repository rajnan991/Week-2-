```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
import re
```

```python
p=pd.read_csv("items.csv")
```

```python
p.head()
```

```python
p.describe()
```

```python
p['product_price']=p['product_price'].str.strip('₹')
p
```

```python
new=p['product_storage'].str.split("|",n=1,expand=True)
p['Ram']=new[0]
p['Rom']=new[1]
p.drop(columns=["product_storage"],inplace=True)
p
```

```python
new=p['Rom'].str.split("|",n=1,expand=True)
p['rom']=new[0]
p['x']=new[1]
p.drop(columns=["Rom"],inplace=True)
p
```

```python
new=p['product_camera'].str.split("|",n=1,expand=True)
p['Camera']=new[0]
p['x']=new[1]
p.drop(columns=["product_camera"],inplace=True)
p
```

```python
new=p['Ram'].str.split(" ",n=1,expand=True)
p['Ram in GB']=new[0]
p['y']=new[1]
p.drop(columns=["Ram"],inplace=True)
p
```

```python
new=p['y'].str.split(" ",n=1,expand=True)
p['Rom in GB']=new[0]
p['a']=new[1]
p.drop(columns=["y"],inplace=True)
p
```

```python
new=p['x'].str.split(" ",n=1,expand=True)
p['Front_Camera']=new[0]
p['a']=new[1]
p.drop(columns=["x"],inplace=True)
p.drop(['Front_Camera'],axis=1,inplace=True)
p
```

```python
new=p['a'].str.split(" ",n=1,expand=True)
p['Front_Camera']=new[0]
p['a']=new[1]
p.drop(columns=["a"],inplace=True)
p
```

```python
p
```

```python
p['Front_Camera']=p['Front_Camera'].str.strip("MP")
p.rename(columns={'Battery': 'Battery in mAh'}, inplace = True)
p.rename(columns={'Back_Camera': 'Back_Camera in MP'}, inplace = True)
p.rename(columns={'Front_Camera': 'Front_Camera in MP'}, inplace = True)
p
```

```python
p.to_csv("newitems.csv")
```
