```python
def clean_dataset(x):
    assert isinstance(x, pd.DataFrame), "h needs to be a pd.DataFrame"
    x.dropna(inplace=True)
    indices_to_keep = ~x.isin([np.nan, np.inf, -np.inf]).any(1)
    return x[indices_to_keep].astype(np.float64)
```
